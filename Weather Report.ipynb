# Performing Analysis Of Meteorological Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "weatherHistory.csv"
file = pd.read_csv(file_path)
```

```python
file.head()
```

```python
titles_req = ["Formatted Date", "Apparent Temperature (C)", "Humidity"]
df = file[titles_req]
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df_1 = df.set_index('Formatted Date')
df_1 = df_1.resample('MS').mean()
```

```python
df_1.head()
```

```python
plt.figure(figsize=(14,6))
plt.title("variation in Apparent Trmperature and Humidity with time")
plt.plot(df_1)
```

```python
df_april = df_1[df_1.index.month==4]
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_april)
plt.xlabel("Years",fontsize=12)
plt.ylabel("Tempereature",fontsize=12)
plt.title("Temperature and Humidity over the year on month of April",fontsize=20)
```
